```python
import pickle
import numpy as np

with open('Data/Models/XGBoost.pkl', 'rb') as f:
    model = pickle.load(f)

print(model.get_booster().feature_names)

# Must match training order exactly
duration_p   = 0.0
wv100_k_av   = 21.13
ny8          = 50.23
ny3_8        = 38.00
f_load_av    = 15.51
f_start      = 0
FWI_12h_av   = 40.23

features = np.array([[
    duration_p,
    wv100_k_av,
    ny8,
    ny3_8,
    f_load_av,
    f_start,
    FWI_12h_av
]])

# model output is log-transformed ROS
prediction_log = model.predict(features)[0]

# invert log transform
inv_mult = np.log(10) / 5
ros = np.exp(prediction_log * inv_mult) - 1

print("ROS (log):", prediction_log)
print("ROS (linear):", ros)
```
