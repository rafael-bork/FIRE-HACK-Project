```python
# --- Quantificar vari√°veis de superf√≠cie (ds_SL) ---
ds_SL["t_2m_K"] = (ds_SL["t_2m_K"] * units.kelvin)
ds_SL["d_2m_K"] = (ds_SL["d_2m_K"] * units.kelvin)

ds_SL["u10_ms"] = (ds_SL["u10_ms"] * (units.meter / units.second))
ds_SL["v10_ms"] = (ds_SL["v10_ms"] * (units.meter / units.second))
ds_SL["u100_ms"] = (ds_SL["u100_ms"] * (units.meter / units.second))
ds_SL["v100_ms"] = (ds_SL["v100_ms"] * (units.meter / units.second))

ds_SL["sP_Pa"] = (ds_SL["sP_Pa"] * units.pascal)
ds_SL["gp_m2s2"] = (ds_SL["gp_m2s2"] * (units.meter ** 2 / units.second ** 2))

# Vari√°veis derivadas de superf√≠cie
ds_SL["Cape"] = ds_SL["Cape"] * units.joule / units.kilogram
ds_SL["Cin"] = ds_SL["Cin"] * units.joule / units.kilogram
ds_SL["CBH_m"] = ds_SL["CBH_m"] * units.meter
ds_SL["BLH_m"] = ds_SL["BLH_m"] * units.meter

# Vari√°veis de cobertura de nuvens
ds_SL["LowCC_p"] = ds_SL["LowCC_p"] * units.percent
ds_SL["MidCC_p"] = ds_SL["MidCC_p"] * units.percent
ds_SL["HigCC_p"] = ds_SL["HigCC_p"] * units.percent
ds_SL["TotCC_p"] = ds_SL["TotCC_p"] * units.percent

# --- Quantificar vari√°veis de n√≠veis de press√£o (ds_PL) ---
ds_PL["t_K"] = ds_PL["t_K"] * units.kelvin
ds_PL["rh"] = ds_PL["rh"] * units.percent
ds_PL["u_ms"] = ds_PL["u_ms"] * (units.meter / units.second)
ds_PL["v_ms"] = ds_PL["v_ms"] * (units.meter / units.second)
ds_PL["w_Pas"] = ds_PL["w_Pas"] * (units.pascal / units.second)
ds_PL["gp"] = ds_PL["gp"] * (units.meter ** 2 / units.second ** 2)

ds_PL["pressure_level"] = ds_PL["pressure_level"] * units.hectopascal
```

```python
# --- Convers√µes de temperatura ---
ds_SL["t_2m_C"] = ds_SL["t_2m_K"].metpy.convert_units("degC")
ds_SL["d_2m_C"] = ds_SL["d_2m_K"].metpy.convert_units("degC")

ds_PL["t_C"] = ds_PL["t_K"].metpy.convert_units("degC")


# --- Convers√µes de vento (10 m) ---
ds_SL["u10_kh"] = ds_SL["u10_ms"].metpy.convert_units("km/h")
ds_SL["v10_kh"] = ds_SL["v10_ms"].metpy.convert_units("km/h")

# --- Convers√µes de vento (100 m) ---
ds_SL["u100_kh"] = ds_SL["u100_ms"].metpy.convert_units("km/h")
ds_SL["v100_kh"] = ds_SL["v100_ms"].metpy.convert_units("km/h")


ds_PL["u_kh"] = ds_PL["u_ms"].metpy.convert_units("km/h")
ds_PL["v_kh"] = ds_PL["v_ms"].metpy.convert_units("km/h")


# --- Convers√£o de press√£o ---
ds_SL["sP_hPa"] = ds_SL["sP_Pa"].metpy.convert_units("hPa")

ds_SL = ds_SL.drop_vars(['sP_Pa', 'u10_ms', 'v10_ms', 'u100_ms', 'v100_ms', 't_2m_K', 'd_2m_K'])
ds_PL = ds_PL.drop_vars(['u_ms', 'v_ms', 't_K'])
```

```python
# --- Velocidade e dire√ß√£o do vento (10 m) ---
ds_SL["wv10_kh"] = np.sqrt(ds_SL["u10_kh"]**2 + ds_SL["v10_kh"]**2)
ds_SL["wdir10"] = (180 + np.degrees(np.arctan2(ds_SL["u10_kh"], ds_SL["v10_kh"]))) % 360

# --- Velocidade e dire√ß√£o do vento (100 m) ---
ds_SL["wv100_k"] = np.sqrt(ds_SL["u100_kh"]**2 + ds_SL["v100_kh"]**2)
ds_SL["wdir100"] = (180 + np.degrees(np.arctan2(ds_SL["u100_kh"], ds_SL["v100_kh"]))) % 360

# --- Velocidade e dire√ß√£o do vento ds_PL ---
ds_PL["wv_kh"] = np.sqrt(ds_PL["u_kh"]**2 + ds_PL["v_kh"]**2)
ds_PL["wdir"] = (180 + np.degrees(np.arctan2(ds_PL["u_kh"], ds_PL["v_kh"]))) % 360
```

```python
# --- VERS√ÉO CORRETA E VALIDADA ---
import numpy as np

# --- Extrair valores ---
p_values = ds_PL["pressure_level"].values * 100  # hPa para Pa (shape: [5])
T_values = ds_PL["t_C"].values + 273.15  # ¬∞C para K (shape: [2486, 5, 61, 41])
w_Pas_values = ds_PL["w_Pas"].values  # Pa/s (shape: [2486, 5, 61, 41])

# --- Reformatar a press√£o para fazer broadcasting ---
# A press√£o precisa ter a mesma dimens√£o que temperatura nos n√≠veis de press√£o
# p_values tem shape (5,) - vamos expandir para (1, 5, 1, 1) para broadcast
p_expanded = p_values.reshape(1, 5, 1, 1)  # agora shape: [1, 5, 1, 1]

# --- C√°lculo da densidade do ar ---
R = 287.0  # J/(kg¬∑K)
rho = p_expanded / (R * T_values)  # kg/m¬≥ (agora funciona o broadcast)

# --- Convers√£o w_Pas (Pa/s) ‚Üí w_ms (m/s) ---
g = 9.80665  # m/s¬≤
w_ms = -w_Pas_values / (rho * g)  # m/s

# --- Converter para km/h ---
w_kh = w_ms * 3.6  # m/s para km/h

# --- Adicionar ao dataset ---
ds_PL["w_kh"] = xr.DataArray(
    data=w_kh,
    dims=ds_PL["w_Pas"].dims,
    coords=ds_PL["w_Pas"].coords,
)

# Remover vari√°vel original
ds_PL = ds_PL.drop_vars(['w_Pas'])

print("Convers√£o feita com sucesso!")
print(f"Shape do resultado: {w_kh.shape}")
print(f"w_kh range: {np.min(w_kh):.2f} to {np.max(w_kh):.2f} km/h")
```

```python
# --- Humidade relativa ---
e_s = mpcalc.saturation_vapor_pressure(ds_SL["t_2m_C"])
e_a = mpcalc.saturation_vapor_pressure(ds_SL["d_2m_C"])
ds_SL["rh_2m"] = mpcalc.relative_humidity_from_dewpoint(ds_SL["t_2m_C"], ds_SL["d_2m_C"]) * 100

# --- Vapour Pressure Deficit ---
vpd = e_s - e_a
vpd_pa = vpd.metpy.convert_units('Pa')

ds_SL['VPD_Pa'] = (('valid_time','latitude','longitude'), vpd_pa.data)
```

```python
# --- Dead Fuel Moisture Content (DFMC) ---
rh_vals = ds_SL["rh_2m"].values 
t_vals = ds_SL["t_2m_C"].values

# Calcula delta e DFMC
delta = np.where(rh_vals < 60, 1, 0)
ds_SL["dfmc"] = (("valid_time", "latitude", "longitude"), 
                 4.37 + 0.161 * rh_vals - 0.1 * (t_vals - 25) - 0.027 * rh_vals * delta)
```

```python
# --- √Ågua no solo ---

ds_SL['sW_1m'] = (ds_SL['sW_7'] * 7 + ds_SL['sW_28'] * 21 + ds_SL['sW_100'] * 72 ) / 100
ds_SL['sW_3m'] = (ds_SL['sW_7'] * 7 + ds_SL['sW_28'] * 21 + ds_SL['sW_100'] * 72 + ds_SL['sW_289'] * 189) / 289
```

```python
# --- C√°lculo do Ponto de Orvalho com Satura√ß√£o sobre √Ågua e Gelo ---
t_C = ds_PL['t_C'].values        
rh = ds_PL['rh'].values

# Prote√ß√£o contra log(0)
rh = np.maximum(rh, 1e-8)

# Satura√ß√£o √°gua/gelo
es_agua = 6.112 * np.exp((17.62 * t_C) / (243.12 + t_C))
es_gelo = 6.112 * np.exp((22.46 * t_C) / (272.62 + t_C))
es = np.where(t_C >= 0, es_agua, es_gelo)

# Press√£o de vapor
e_hPa = rh/100 * es

# Ponto de orvalho
alpha = np.log(e_hPa / 6.112)
Td = np.where(
    t_C >= 0,
    (243.12 * alpha)/(17.62 - alpha),
    (272.62 * alpha)/(22.46 - alpha)
)

# Atualizar Dataset com Dask
ds_PL['d_C'] = (ds_PL['t_C'].dims, Td)
ds_PL['d_C'] = ds_PL['t_C']
```

```python
# --- Negative Convective Inhibition ---
ds_SL['Cin'] = -1 * ds_SL['Cin']

# --- Lifted condensation level (LCL) ---

lcl_pressure, lcl_temperature = mpcalc.lcl(ds_SL["sP_hPa"], ds_SL["t_2m_C"], ds_SL["d_2m_C"])
ds_SL["LCL_hPa"] = (("valid_time", "latitude", "longitude"), lcl_pressure.to(units.hPa).magnitude)

ds_SL['LCL_m'] = mpcalc.pressure_to_height_std(ds_SL["LCL_hPa"] * units.hPa)

# --- Hot Dry Windy Index (HDW) ---

ds_SL["HDW"] = ds_SL['VPD_Pa'] * ds_SL['wv10_kh']
```

```python
# Constante gravidade
g = 9.80665 * units('m/s^2')

# Fun√ß√£o Haines Index CORRIGIDA
def haines_index_pixel(T, Td, level):
    if level == 'low':
        p1, p2 = 950, 850
        lapse_thresh = [4, 8]
        dew_thresh = [6, 12]
    elif level == 'mid':
        p1, p2 = 850, 700
        lapse_thresh = [6, 11]
        dew_thresh = [6, 13]
    else:  # high
        p1, p2 = 700, 500
        lapse_thresh = [11, 17]
        dew_thresh = [15, 21]

    # Selecionar os n√≠veis de press√£o
    T1 = T.sel(pressure_level=p1)
    T2 = T.sel(pressure_level=p2)
    Td2 = Td.sel(pressure_level=p2)

    # Calcular lapse rate e deficit de ponto de orvalho
    lapse = (T1 - T2)
    dew_deficit = (T2 - Td2)

    lapse_vals = lapse.metpy.magnitude
    dew_vals = dew_deficit.metpy.magnitude

    # Aplicar condi√ß√µes com tratamento de NaN
    lapse_score = xr.where(np.isnan(lapse_vals), np.nan,
                    xr.where(lapse_vals < lapse_thresh[0], 1,
                    xr.where(lapse_vals <= lapse_thresh[1], 2, 3)))

    dew_score = xr.where(np.isnan(dew_vals), np.nan,
                  xr.where(dew_vals < dew_thresh[0], 1,
                  xr.where(dew_vals <= dew_thresh[1], 2, 3)))

    return lapse_score + dew_score

# Converter geopotencial da superf√≠cie em altitude
z_sfc = ds_SL['gp_m2s2'] / g

# Criar m√°scara de n√≠vel do Haines
level_array = xr.where(z_sfc < 1000 * units.meter, 'low',
               xr.where(z_sfc < 3000 * units.meter, 'mid', 'high'))

# Inicializar DataArray para Haines Index
haines = xr.full_like(z_sfc, fill_value=np.nan)

# Calcular Haines Index para cada n√≠vel
for lvl in ['low','mid','high']:
    mask = level_array == lvl
    haines = xr.where(mask, haines_index_pixel(ds_PL['t_C'],
                                               ds_PL['d_C'], lvl), haines)

# Adicionar ao ds_SL
ds_SL['Haines'] = haines
```

```python
# --- Altura da superf√≠cie (geopotential ‚Üí height) ---
z_sfc = (
    ds_SL['gp_m2s2']
    .metpy.quantify()
    .metpy.convert_units('meter**2/second**2')
) / (9.80665 * units('meter/second**2'))
z_sfc = z_sfc.metpy.convert_units('meter')

# --- Usa vento e dire√ß√£o j√° calculados no Dataset ---
ws_sfc = ds_SL['wv10_kh'].metpy.dequantify()   # vento 10 m
dir_sfc = ds_SL['wdir10'].metpy.dequantify()   # dire√ß√£o 10 m

# --- Loop nos n√≠veis ---
pressure_levels = [950, 700, 500]

for level in pressure_levels:
    # Pressure (com unidades)
    p = ds_PL['pressure_level'].sel(pressure_level=level).values * units.hPa

    # Altura padr√£o para o n√≠vel
    z_pl = mpcalc.pressure_to_height_std(p).to('meter')

    # Espessura em km (altura do n√≠vel - altura da superf√≠cie)
    dz_km = (z_pl - z_sfc) / (1000 * units.meter)

    # Vento e dire√ß√£o no n√≠vel (j√° calculados)
    ws_pl = ds_PL['wv_kh'].sel(pressure_level=level).metpy.dequantify()
    dir_pl = ds_PL['wdir'].sel(pressure_level=level).metpy.dequantify()

    # Shear de magnitude e dire√ß√£o
    shear_mag = (ws_pl - ws_sfc) / dz_km
    d_dir = ((dir_pl - dir_sfc + 180) % 360) - 180
    shear_dir = d_dir / dz_km

    # Identificador do n√≠vel
    level_first_digit = str(level)[0]

    # Salva no Dataset
    ds_SL[f'wSv_{level_first_digit}'] = (('valid_time', 'latitude', 'longitude'), shear_mag.data)
    ds_SL[f'wSdir_{level_first_digit}'] = (('valid_time', 'latitude', 'longitude'), shear_dir.data)

# --- Shear entre 10 m e 100 m ---
ws_100 = ds_SL['wv100_k'].metpy.dequantify()
dir_100 = ds_SL['wdir100'].metpy.dequantify()

shear_mag_10_100 = (ws_100 - ws_sfc) / 0.09  # 90 m = 0.09 km
d_dir_10_100 = ((dir_100 - dir_sfc + 180) % 360) - 180
shear_dir_10_100 = d_dir_10_100 / 0.09

# Adiciona ao Dataset
ds_SL['wSv_1'] = (('valid_time', 'latitude', 'longitude'), shear_mag_10_100.data)
ds_SL['wSdir_1'] = (('valid_time', 'latitude', 'longitude'), shear_dir_10_100.data)
```

```python
def calculate_lfc_vectorized(surface_pressure, surface_temp, surface_dewpoint, pressure_levels, temps, dews):
    results = np.full(surface_pressure.shape, np.nan)
    
    for i in np.ndindex(surface_pressure.shape):
        try:
            sP = surface_pressure[i]
            t2m = surface_temp[i]
            d2m = surface_dewpoint[i]
            
            # Se algum dado de entrada for NaN, mant√©m NaN
            if np.isnan(sP) or np.isnan(t2m) or np.isnan(d2m):
                results[i] = np.nan
                continue
                
            t_prof = temps[:, i[0], i[1]]
            d_prof = dews[:, i[0], i[1]]
            
            # Verificar se h√° dados suficientes no perfil
            if np.all(np.isnan(t_prof)) or np.all(np.isnan(d_prof)):
                results[i] = -1  # Perfil insuficiente
                continue
            
            full_pressure = np.concatenate([[sP], pressure_levels])
            full_temperature = np.concatenate([[t2m], t_prof])
            full_dewpoint = np.concatenate([[d2m], d_prof])
            
            # Remover NaNs do perfil
            mask = ~(np.isnan(full_pressure) | np.isnan(full_temperature) | np.isnan(full_dewpoint))
            if np.sum(mask) < 3:  # Muito poucos pontos
                results[i] = -1
                continue
                
            full_pressure = full_pressure[mask]
            full_temperature = full_temperature[mask]
            full_dewpoint = full_dewpoint[mask]
            
            sort_idx = np.argsort(full_pressure)[::-1]
            pressure_sorted = full_pressure[sort_idx] * units.hPa
            temp_sorted = full_temperature[sort_idx] * units.degC
            dewpoint_sorted = full_dewpoint[sort_idx] * units.degC
            
            lfc_pressure, lfc_temp = mpcalc.lfc(pressure_sorted, temp_sorted, dewpoint_sorted)
            
            if lfc_pressure is not None and not np.isnan(lfc_pressure.magnitude):
                results[i] = lfc_pressure.magnitude
            else:
                results[i] = -1  # LFC meteorologicamente n√£o definido

        except Exception as e:
            results[i] = -1  # Erro no c√°lculo
    
    return results
```

```python
def calculate_ccl(surface_pressure, surface_temp, surface_dewpoint, pressure_levels, temps, dews):
    # VERS√ÉO ORIGINAL
    results = np.full(surface_pressure.shape, np.nan)
    
    for i in np.ndindex(surface_pressure.shape):
        try:
            sP = surface_pressure[i]
            t2m = surface_temp[i]
            d2m = surface_dewpoint[i]
            
            # Se algum dado de entrada for NaN, mant√©m NaN
            if np.isnan(sP) or np.isnan(t2m) or np.isnan(d2m):
                results[i] = np.nan
                continue
                
            t_prof = temps[:, i[0], i[1]]
            d_prof = dews[:, i[0], i[1]]
            
            # Verificar se h√° dados v√°lidos no perfil
            if np.all(np.isnan(t_prof)) or np.all(np.isnan(d_prof)):
                results[i] = -1  # Dados do perfil insuficientes
                continue
            
            full_pressure = np.concatenate([[sP], pressure_levels])
            full_temperature = np.concatenate([[t2m], t_prof])
            full_dewpoint = np.concatenate([[d2m], d_prof])
            
            # Remover NaNs do perfil
            mask = ~(np.isnan(full_pressure) | np.isnan(full_temperature) | np.isnan(full_dewpoint))
            if np.sum(mask) < 3:  # Muito poucos pontos
                results[i] = -1
                continue
                
            full_pressure = full_pressure[mask]
            full_temperature = full_temperature[mask]
            full_dewpoint = full_dewpoint[mask]
            
            sort_idx = np.argsort(full_pressure)[::-1]
            pressure_sorted = full_pressure[sort_idx] * units.hPa
            temp_sorted = full_temperature[sort_idx] * units.degC
            dewpoint_sorted = full_dewpoint[sort_idx] * units.degC
            
            ccl_pressure, ccl_temp, conv_temp = mpcalc.ccl(pressure_sorted, temp_sorted, dewpoint_sorted)
            
            if ccl_pressure is not None and not np.isnan(ccl_pressure.magnitude):
                results[i] = ccl_pressure.magnitude
            else:
                results[i] = -1  # C√°lculo meteorol√≥gico n√£o definido
                
        except Exception as e:
            results[i] = -1  # Erro no c√°lculo
    
    return results
```

```python
def calculate_equilibrium_level(surface_pressure, surface_temp, surface_dewpoint, pressure_levels, temps, dews, cape):
    results = np.full(surface_pressure.shape, np.nan)
    
    for i in np.ndindex(surface_pressure.shape):
        try:
            sP = surface_pressure[i]
            t2m = surface_temp[i]
            d2m = surface_dewpoint[i]
            cape_val = cape[i]
            
            # Se algum dado de entrada for NaN, mant√©m NaN
            if np.isnan(sP) or np.isnan(t2m) or np.isnan(d2m) or np.isnan(cape_val):
                results[i] = np.nan
                continue
                
            t_prof = temps[:, i[0], i[1]]
            d_prof = dews[:, i[0], i[1]]
            
            # Verificar se h√° dados suficientes no perfil
            if np.all(np.isnan(t_prof)) or np.all(np.isnan(d_prof)):
                results[i] = -1  # Perfil insuficiente
                continue
            
            pressure = np.concatenate([[sP], pressure_levels])
            temperature = np.concatenate([[t2m], t_prof])
            dewpoint = np.concatenate([[d2m], d_prof])
            
            # Remover NaNs do perfil
            mask = ~(np.isnan(pressure) | np.isnan(temperature) | np.isnan(dewpoint))
            if np.sum(mask) < 3:  # Muito poucos pontos
                results[i] = -1
                continue
                
            pressure = pressure[mask] * units.hPa
            temperature = temperature[mask] * units.degC
            dewpoint = dewpoint[mask] * units.degC
            
            # Ordenar
            sort_idx = np.argsort(pressure.m)[::-1]
            pressure = pressure[sort_idx]
            temperature = temperature[sort_idx]
            dewpoint = dewpoint[sort_idx]
            
            # Verificar se h√° CAPE (condi√ß√£o necess√°ria para EL)
            if cape_val <= 0:
                results[i] = -1  # Sem CAPE, n√£o h√° EL
                continue
            
            # Usar a fun√ß√£o espec√≠fica do MetPy para encontrar EL
            el_pressure, el_temperature = mpcalc.el(pressure, temperature, dewpoint)
            
            if el_pressure is not None and not np.isnan(el_pressure.magnitude):
                results[i] = mpcalc.pressure_to_height_std(el_pressure).m
            else:
                results[i] = -1  # EL meteorologicamente n√£o definido
                
        except:
            results[i] = -1  # Erro no c√°lculo
    
    return results
```

```python
import numpy as np

def calculate_ventilation_index_mlh_950_vectorized(
        surface_pressure,     # (lat, lon)
        surface_temp,         # (lat, lon)  
        surface_dewpoint,     # (lat, lon)
        u950,                 # (lat, lon) - J√Å CORRETO!
        v950,                 # (lat, lon)
        cape):                # (lat, lon)
    
    # -------- 1. WIND SPEED NA 950 hPa --------
    wind_speed_950 = np.sqrt(u950**2 + v950**2)

    # -------- 2. DEWPOINT DEPRESSION --------
    dewpoint_dep = surface_temp - surface_dewpoint

    # -------- 3. MIXING LAYER HEIGHT (MLH) --------
    mlh = np.full_like(surface_temp, np.nan, dtype=float)

    # Condi√ß√µes convectivas
    mask1 = (cape > 100) & (wind_speed_950 < 10)
    mlh[mask1] = 1500.0

    # Condi√ß√µes ventosas
    mask2 = (cape < 50) & (wind_speed_950 > 15)
    mlh[mask2] = 800.0

    # Muito est√°vel
    mask3 = cape < 10
    mlh[mask3] = 500.0

    # Condi√ß√µes neutras baseadas no spread T - Td
    mask4 = ~ (mask1 | mask2 | mask3)
    mlh_neutral = np.full_like(surface_temp, np.nan, dtype=float)

    mlh_neutral[(dewpoint_dep > 15) & mask4] = 1200.0
    mlh_neutral[(dewpoint_dep > 8)  & (dewpoint_dep <= 15) & mask4] = 1000.0
    mlh_neutral[(dewpoint_dep <= 8) & mask4] = 700.0

    mlh[mask4] = mlh_neutral[mask4]

    # -------- 4. CALCULAR VENTILATION INDEX --------
    ventilation = wind_speed_950 * mlh

    # Onde entradas s√£o NaN ‚Üí sa√≠da vira NaN automaticamente
    ventilation[np.isnan(wind_speed_950)] = np.nan
    ventilation[np.isnan(cape)] = np.nan
    ventilation[np.isnan(surface_temp)] = np.nan
    ventilation[np.isnan(surface_dewpoint)] = np.nan

    return ventilation
```

```python
def calculate_lifted_index(surface_pressure, surface_temp, surface_dewpoint, pressure_levels, temps, dews):
    results = np.full(surface_pressure.shape, np.nan)
    
    for i in np.ndindex(surface_pressure.shape):
        sP = surface_pressure[i]
        t2m = surface_temp[i]
        d2m = surface_dewpoint[i]
        t_prof = temps[:, i[0], i[1]]
        d_prof = dews[:, i[0], i[1]]
        
        try:
            if np.isnan(sP) or np.isnan(t2m) or np.isnan(d2m):
                results[i] = np.nan
                continue
            
            pressure_500_idx = np.where(pressure_levels == 500)[0]
            if len(pressure_500_idx) == 0:
                results[i] = -1
                continue
                
            t_500 = t_prof[pressure_500_idx[0]]
            
            if np.isnan(t_500):
                results[i] = -1
                continue
            
            t2m_kelvin = (t2m + 273.15) * units.kelvin
            
            parcel_profile_500 = mpcalc.dry_lapse(
                np.array([sP, 500]) * units.hPa, 
                t2m_kelvin
            )
            
            t_parcel_500 = parcel_profile_500[-1].m - 273.15
            li = t_500 - t_parcel_500
            
            results[i] = li
            
        except Exception as e:
            results[i] = -1
    
    return results
```

```python
# Calcular todos os par√¢metros
print("Calculando par√¢metros adicionais...")

total_times = len(ds_SL.valid_time)

# Arrays para os novos par√¢metros
lfc_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
ccl_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
equilibrium_level_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
ventilation_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
li_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)

SBI_i_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
SBI_t_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
SBI_g_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)

EBI_i_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
EBI_t_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
EBI_g_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)
EBI_base_p_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)

multiple_EBIs_array = np.full((total_times, len(ds_SL.latitude), len(ds_SL.longitude)), np.nan)


for time_idx in range(total_times):

    print(f"Processando {time_idx + 1}/{total_times} tempos ({(time_idx + 1)/total_times*100:.1f}%)")

    surface_data = ds_SL.isel(valid_time=time_idx)
    profile_data = ds_PL.isel(valid_time=time_idx)

    # LFC
    lfc_array[time_idx] = calculate_lfc_vectorized(
        surface_data['sP_hPa'].values,
        surface_data['t_2m_C'].values,
        surface_data['d_2m_C'].values,
        ds_PL.pressure_level.values,
        profile_data['t_C'].values,
        profile_data['d_C'].values
    )

    # CCL
    ccl_array[time_idx] = calculate_ccl(
        surface_data['sP_hPa'].values,
        surface_data['t_2m_C'].values,
        surface_data['d_2m_C'].values,
        ds_PL.pressure_level.values,
        profile_data['t_C'].values,
        profile_data['d_C'].values
    )
    
    # Cloud Top Height
    equilibrium_level_array[time_idx] = calculate_equilibrium_level(
        surface_data['sP_hPa'].values,
        surface_data['t_2m_C'].values,
        surface_data['d_2m_C'].values,
        ds_PL.pressure_level.values,
        profile_data['t_C'].values,
        profile_data['d_C'].values,
        surface_data['Cape'].values
    )

    # Ventilation Index
    idx_950 = np.where(ds_PL.pressure_level.values == 950)[0][0]
    u950 = ds_PL['u_kh'].isel(pressure_level=idx_950, valid_time=time_idx).values
    v950 = ds_PL['v_kh'].isel(pressure_level=idx_950, valid_time=time_idx).values

    ventilation_array[time_idx] = calculate_ventilation_index_mlh_950_vectorized(
        surface_data['sP_hPa'].values,
        surface_data['t_2m_C'].values,
        surface_data['d_2m_C'].values,
        u950,
        v950,
        surface_data['Cape'].values
    )


    # Lifted Index
    li_array[time_idx] = calculate_lifted_index(
        surface_data['sP_hPa'].values,
        surface_data['t_2m_C'].values,
        surface_data['d_2m_C'].values,
        ds_PL.pressure_level.values,
        profile_data['t_C'].values,
        profile_data['d_C'].values
    )

#    if time_idx == 100:
#        break

# Adicionar todos ao dataset
ds_SL['LFC_hPa'] = (('valid_time', 'latitude', 'longitude'), lfc_array)
ds_SL['CCL_hPa'] = (('valid_time', 'latitude', 'longitude'), ccl_array)
ds_SL['EL_m'] = (('valid_time', 'latitude', 'longitude'), equilibrium_level_array)
ds_SL['VentIdx'] = (('valid_time', 'latitude', 'longitude'), ventilation_array)
ds_SL['LiftIdx'] = (('valid_time', 'latitude', 'longitude'), li_array)

print("C√°lculo de todos os par√¢metros conclu√≠do!")
```
