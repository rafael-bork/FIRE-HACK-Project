```python
import geopandas as gpd
import pandas as pd

# Leitura do shapefile
df = gpd.read_file(r"Data\Processed\PT-FireSprd_v2.1\L2_FireBehavior\PT-FireProg_v2.1_L2_final_with_SHAP.shp")

# Converter sdate para datetime
df['sdate'] = pd.to_datetime(df['sdate'])

# Definir a data que queremos filtrar (somente ano-mês-dia)
data_filtrar = pd.to_datetime("2017-10-15")

# Filtrar pelo PredROS existente e sdate específico
df_filtrado = df[(df['PredROS'].notna()) & (df['sdate'].dt.date == data_filtrar.date())]

# Mostrar os resultados
print(df_filtrado.head())

df_filtrado.to_csv(r"Data\Processed\PT-FireSprd_v2.1\L2_FireBehavior\PT-FireProg_v2.1_L2_filtrado.csv", index=False)

```
